```python
def analyze_numbers():
  while True:
    try:
      count_str = input("How many numbers do you want to enter?: ")
      num_count = int(count_str)
      if num_count <= 0:
        print("Please enter a positive number.")
        continue
      break
    except ValueError:
      print("Invalid input. Please enter a whole number.")

  total_sum = 0
  valid_numbers_entered = 0

  print(f"\nPlease enter {num_count} numbers:")

  for i in range(num_count):
    while True:
      num_input = input(f"Enter number {i + 1}: ")
      try:
        num = int(num_input)
        if num % 2 == 0:
          print(f"{num} is even.")
        else:
          print(f"{num} is odd.")
        total_sum += num
        valid_numbers_entered += 1
        break
      except ValueError:
        print("Invalid input. Please enter a whole number.")

  print("\n--- Analysis Results ---")
  if valid_numbers_entered > 0:
    average = total_sum / valid_numbers_entered
    print(f"Total sum of numbers: {total_sum}")
    print(f"Average of numbers: {average:.2f}")
  else:
    print("No valid numbers were entered.")

  print("------------------------")
```

```python
analyze_numbers()
```
